```python
import pandas as pd

veriler = pd.read_csv("customer_data.csv")
veriler.head()
```

```python
import matplotlib as plt
import seaborn as sns
sns.pairplot(veriler)
```

```python
korelasyon_matrisi = veriler.corr()
```

```python
heatmap = sns.heatmap(veriler.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
X = veriler['income'].values
y = veriler['purchase_amount'].values
```

```python
uzunluk = len(X)
X = X.reshape((uzunluk,1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
lm = LinearRegression()
lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, lm.predict(X_train), color = 'red')
plt.title('Gelir düzeyine göre satış miktarı')
plt.xlabel('Gelir')
plt.ylabel('satış miktarı')
plt.show()
```

```python
print(lm.score(X_test, y_test))
```
